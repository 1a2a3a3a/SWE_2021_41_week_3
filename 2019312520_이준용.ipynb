```python
def isHappy(n):
  results =[];
  sign=0
  while(1 not in results):
    square_sums=0
    while(n>0):
      m=n%10
      n//=10
      square_sums+=m**2
    if (square_sums in results):
      sign=1
      break
    else:
      results.append(square_sums)
      n=square_sums

  if sign==0:
    return True
  else:
    return False

```
